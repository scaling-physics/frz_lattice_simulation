```python
def moving_average(lst, window_size):
    """
    Calculate the moving average of a list.

    Args:
        lst (list): The input list of values.
        window_size (int): The window size for the moving average.

    Returns:
        list: A list containing the moving average values.
    """
    return [sum(lst[i-window_size+1:i+1])/len(lst[i-window_size+1:i+1]) if i >= window_size-1 else sum(lst[0:i+1])/(i+1) for i in range(len(lst))]
```

```python
import imageio
import os
from natsort import natsorted
alpha = 0
density = 0.1
run = 9
off = 0.03

for index in [5]:
    for J in [4]:
        for beta in [2*J]:
            for FrzB in [35]:
                # Directory containing the input images
                input_dir = f'/scratch2/Srikanth/chemotactic_clusters/{run}/{index}_dFrzB_labels_J_{J}_alpha_{alpha}_FrzB_{FrzB}_beta_{beta}'

                # Get a list of all image file names in the directory
                image_files = natsorted([f for f in os.listdir(input_dir) if f.endswith('.jpg') or f.endswith('.png')])

                # Create a list to hold the image data
                image_data = []

                # Iterate through the input images and read them
                for image_file in image_files:
                    image_path = os.path.join(input_dir, image_file)
                    image = imageio.imread(image_path)
                    image_data.append(image)

                # Save the image data as a GIF
                output_path = input_dir+'/output.gif'  # Output GIF file name
                fps = 10 # Frames per second for the output GIF
                imageio.mimsave(output_path, image_data, fps=fps)
```
